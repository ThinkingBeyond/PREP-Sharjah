## Input

Specify all the variables

- position `pos` (type: 3D vector) as a `numpy` array
- time `t` (type: real scalar)
  - initial time `t_0` ($t_0$)
  - time step `dt` ($\Delta t$)
  - number of time steps `N`

```python
# Python library for fast numerical computations with vectors
import numpy as np

pos = np.array([0,0,0])
t_0 = 0
dt = 0.01
N = 1000
```

## Run the Simulation

Here is where we do all the necessary computation

**Plan**

*We want:* Compute motion as a function $\gamma: \mathbb{R}\to \mathbb{R}^3$

1. Define the function of motion we want to simulate
2. Create a list (or an array) of positions for each time step.
    1. Start at given initial time $t_0$
    2. Calling $\gamma(t_0)$
    3. Add position to list
    4. Update time to $t_1 = t_0 +\Delta t$
    5. Calling $\gamma(t_1)$
    6. Repeat this for the number of time steps

**Implementation**

1. Define a function `motion` with input time `t`
2. Run a FOR loop over the number of time steps from 0 to `N`


```python
# Calculates the postion at time t for a uniform linear motion
def motion(t):
  # Set initial position
  initial_pos = np.array([.0,.0,.0])
  # Set initial velocity
  v_0 = np.array([1.0,2.0,3.0])
  pos = initial_pos + t*v_0
  return pos

# Initialise time with t_0
t = t_0

# Initialise trajectory
trajectory = []

# Calculate postions for each point in time and store them in a list
for i in range(N+1):
    pos = motion(t)
    t = t + dt
    trajectory.append(pos)

# Test we got a list
print(trajectory)
```

## Generate the Output

Plot the simulated trajectory in 3D.

```python
# Imports
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- 2. Process Data for Matplotlib ---

# Matplotlib's plot function expects separate arrays for X, Y, and Z coordinates.
# We convert the list of arrays into a single (N, 3) NumPy array,
# and then split it into columns.
trajectory_matrix = np.array(trajectory)

# Extract coordinates:
X = trajectory_matrix[:, 0]
Y = trajectory_matrix[:, 1]
Z = trajectory_matrix[:, 2]

# --- 3. Plotting the Trajectory ---

# Create a figure and a 3D Axes object
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the path (the line connecting all points)
# 'r-' specifies a red, solid line.
ax.plot(X, Y, Z, color='r', label='3D Trajectory Path', linewidth=2)

# Optionally, plot the start and end points
ax.scatter(X[0], Y[0], Z[0], color='g', marker='x', s=100, label='Start Point')
ax.scatter(X[-1], Y[-1], Z[-1], color='b', marker='x', s=100, label='End Point')

# --- 4. Set Labels and Title ---

ax.set_xlabel('X Coordinate (Units)')
ax.set_ylabel('Y Coordinate (Units)')
ax.set_zlabel('Z Coordinate (Units)')
ax.set_title('3D Path of a Simulated Trajectory')
ax.legend()

# Set plot limits to ensure equal aspect ratio (optional, but often helpful in 3D)
# Setting aspect='equal' is tricky in 3D, so we usually rely on `ax.set_box_aspect`
# to ensure the path doesn't look stretched, but we won't strictly enforce it here.
```
